```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/gender_submission.csv')
```

```python
df['Embarked'] = np.random.choice(['S', 'C', 'Q'], size=len(df), p=[0.7, 0.2, 0.1])
df['Sex'] = np.random.choice(['male', 'female'], size=len(df), p=[0.6, 0.4])
```

```python
ports = pd.get_dummies(df.Embarked , prefix='Embarked')
ports.head()
```

```python
df=df.join(ports)
df.drop(['Embarked'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.Sex = df.Sex.map({'male':0, 'female':1})
```
